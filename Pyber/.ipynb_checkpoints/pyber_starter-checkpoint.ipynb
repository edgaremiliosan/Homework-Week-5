```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
combined_df = pd.merge(ride_df,city_df, on="city", how="left")

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
bubble_plot_df = combined_df.set_index('type')

rural_x_axis = bubble_plot_df.loc['Rural','driver_count'].sum()
urban_x_axis = bubble_plot_df.loc['Urban','driver_count'].sum()
suburban_x_axis = bubble_plot_df.loc['Suburban','driver_count'].sum()
rural_y_axis = bubble_plot_df.loc['Rural','fare'].mean()
urban_y_axis = bubble_plot_df.loc['Urban','fare'].mean()
suburban_y_axis = bubble_plot_df.loc['Suburban','fare'].mean()

# Build the scatter plots for each city types
plt.scatter(rural_x_axis, rural_y_axis, marker="o", facecolors="lightcoral", edgecolors="black",
            s=rural_x_axis, alpha=0.75)
plt.scatter(urban_x_axis, urban_y_axis, marker="o", facecolors="lightskyblue", edgecolors="black",
            s=rural_x_axis, alpha=0.75)
plt.scatter(suburban_x_axis, suburban_y_axis, marker="o", facecolors="gold", edgecolors="black",
            s=rural_x_axis, alpha=0.75)
# Incorporate the other graph properties

# Create a legend
# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_counts = combined_df['fare'].sum()
fare_pie = combined_df.groupby(['type'])
fare_totals = fare_pie['fare'].sum()

percentage_calc =  fare_totals / fare_counts * 100

percentage_df = pd.DataFrame({'Percentage': percentage_calc})
rural_percent = str(percentage_df.loc['Rural','Percentage'])
suburban_percent = str(percentage_df.loc['Suburban','Percentage'])
urban_percent = str(percentage_df.loc['Urban','Percentage'])

percentage = [rural_percent,suburban_percent,urban_percent]
types = ['Rural','Suburban','Urban']
colors = ['gold','lightskyblue','lightcoral']
explode = [0,0,0.1]

# Build Pie Chart
plt.pie(percentage, explode=explode, labels=types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
plt.show()
# Save Figure
plt.savefig("plots/Total Fares by City Type.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_counts = combined_df['ride_id'].count()
ride_pie = combined_df.groupby(['type'])
ride_totals = ride_pie['ride_id'].count()

ride_percentage_calc =  ride_totals / ride_counts * 100

percentage_df = pd.DataFrame({'Percentage': ride_percentage_calc})
rural_percent = percentage_df.loc['Rural','Percentage']
suburban_percent = percentage_df.loc['Suburban','Percentage']
urban_percent = percentage_df.loc['Urban','Percentage']

percentage = [rural_percent,suburban_percent,urban_percent]
types = ['Rural','Suburban','Urban']
colors = ['gold','lightskyblue','lightcoral']
explode = [0,0,0.1]

# Build Pie Chart
plt.pie(percentage, explode=explode, labels=types, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")
plt.show()

plt.savefig("plots/Total Rides by City Type.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_counts = combined_df['driver_count'].sum()
driver_pie = combined_df.groupby(['type'])
driver_totals = driver_pie['driver_count'].sum()

driver_percentage_calc =  driver_totals / driver_counts * 100

percentage_df = pd.DataFrame({'Percentage': driver_percentage_calc})
rural_percent = percentage_df.loc['Rural','Percentage']
suburban_percent = percentage_df.loc['Suburban','Percentage']
urban_percent = percentage_df.loc['Urban','Percentage']

percentage = [rural_percent,suburban_percent,urban_percent]
types = ['Rural','Suburban','Urban']
colors = ['gold','lightskyblue','lightcoral']
explode = [0,0,0.2]

# Build Pie Chart
plt.pie(percentage, explode=explode, labels=types, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=170)
plt.title("% of Total Drivers by City Type")
plt.show()

plt.savefig("plots/Total Drivers by City Type.png")
```
